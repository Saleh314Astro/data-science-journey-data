```python
# ── The Daily Ritual ────────────────────────────────────────────
# Load. Look. Clean. Then think. The order never changes.

import pandas as pd
import matplotlib.pyplot as plt

base = "https://raw.githubusercontent.com/Saleh314Astro/data-science-journey-data/main/clean/"
temp = pd.read_csv(base + "gistemp_monthly.csv", parse_dates=["date"]).rename(columns={"value": "anomaly"}).dropna()  # NASA temperature
co2  = pd.read_csv(base + "keeling_co2_monthly.csv", parse_dates=["date"]).rename(columns={"value": "co2"}).dropna()    # Mauna Loa CO2
co2.head()
```

### Stretch lane — does CO2 lead or follow?

If CO2 drives temperature, maybe the temperature responds a little later. Shift the CO2 series by a number of months and see which shift makes the correlation strongest.

</div>

```python
# Stretch lane — shift CO2 in time and re-measure r
for lag in [-120, -60, 0, 60, 120]:
    r_lag = merged["co2"].shift(lag).corr(merged["anomaly"])
    print(f"CO2 shifted {lag:+4d} months -> r = {r_lag:.3f}")
# Which shift wins? Does a shift of a few years change the story much? What does that tell you
# about using correlation alone to argue about cause and timing?
```
