```python
# ── The Daily Ritual ────────────────────────────────────────────
# Load. Look. Clean. Plot. Then think. The order never changes.

import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/Saleh314Astro/data-science-journey-data/main/clean/gistemp_monthly.csv"
df = pd.read_csv(url, parse_dates=["date"])   # cheat sheet P-1
df = df.rename(columns={"value": "anomaly"})  # published column is named "value"
df = df.dropna()                              # every dataset gets the same greeting
df.head()
```

```python
# First reveal: 146 years, month by month

fig, ax = plt.subplots(figsize=(11, 4))
ax.plot(df["date"], df["anomaly"], color="#7B5EA7", linewidth=0.6)
ax.axhline(0, color="#2A2440", linewidth=0.8)
ax.set_xlabel("Year")
ax.set_ylabel("Temperature anomaly (°C)")
ax.set_title("NASA GISTEMP — global monthly anomalies, 1880–present")
plt.show()
```

### Prediction 2 — the journalist's window

A well-known journalist wrote: <b>"Global warming stopped in 1998 — and the data proves it."</b> He chose to show his readers the period 1998–2012 only.

```python
# The journalist's window: 1998–2012

window = df[(df["date"] >= "1998-01-01") & (df["date"] <= "2012-12-31")]

fig, ax = plt.subplots(figsize=(11, 4))
ax.plot(window["date"], window["anomaly"], color="#2f9e94", linewidth=1.0)
ax.set_xlabel("Year")
ax.set_ylabel("Temperature anomaly (°C)")
ax.set_title("1998–2012")
plt.show()
```
