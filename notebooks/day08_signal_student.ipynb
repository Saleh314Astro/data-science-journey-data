```python
# ── The Daily Ritual ────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base = "https://raw.githubusercontent.com/Saleh314Astro/data-science-journey-data/main/clean/"
df = pd.read_csv(base + "gistemp_monthly.csv", parse_dates=["date"]).rename(columns={"value": "anomaly"}).dropna()
df["year"] = df["date"].dt.year + (df["date"].dt.month - 1) / 12
df["x"] = df["year"] - 2000        # centered year: keeps the fitting numerically clean
df.head()
```

```python
# Three fits on the SAME past — three very different futures
future = np.linspace(-120, 100, 400)      # centered years, 1880 -> 2100
fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(df["x"], df["anomaly"], color="lightgray", linewidth=0.5, label="the data")
for deg, color in [(1, "tab:blue"), (3, "tab:orange"), (7, "tab:red")]:
    c = np.polyfit(df["x"], df["anomaly"], deg)
    ax.plot(future, np.polyval(c, future), color=color, label=f"degree {deg}")
    print(f"degree {deg}: 2100 -> {np.polyval(c, 100):+.1f} degC")
ax.set_ylim(-3, 8)                        # clip: degree 7 leaves the planet
ax.set_xlabel("centered year  (0 = 2000)")
ax.set_ylabel("Temperature anomaly (degC)")
ax.set_title("One past, three futures")
ax.legend()
plt.show()
```

### The turn — the data has run out of answers

The numbers alone cannot tell you 2100. Only **physics** can. Here is the mystery planted on Day 6:

> From sunlight alone, what temperature *should* the Earth be?

- **Name:** …
- **My guess (in degC):** …

</div>

<div dir="rtl" style="background:#f5f2fb; border-right: 4px solid #7B5EA7; padding: 8px 14px; border-radius: 6px; font-size: 0.95em; color:#2A2440;">
<b>ميزان الطاقة</b> <span dir="ltr">(energy balance)</span>: كوكب متّزن يُشعّ إلى الفضاء كل ما يستقبله من الشمس. الطاقة الداخلة <span dir="ltr">S(1−α)/4</span> تساوي الخارجة <span dir="ltr">σT⁴</span>. حُلّ للحرارة <span dir="ltr">T</span> فتحصل على حرارة الصخر العاري بلا غلاف جوي.
</div>

```python
# The physics the data could not give you: a bare rock in the Sun
S = 1361.0        # solar constant, W/m2 (sunlight arriving at Earth's distance)
albedo = 0.30     # fraction reflected straight back (clouds, ice, deserts)
sigma = 5.67e-8   # Stefan-Boltzmann constant

T = (S * (1 - albedo) / (4 * sigma)) ** 0.25
print("Bare-rock Earth should be:", round(T, 1), "K =", round(T - 273.15, 1), "degC")
print("Measured Earth: about +15 degC")
print("The gap:", round(15 - (T - 273.15), 1), "degC  <-- this is the greenhouse effect")
```

### Stretch lane — test the fits on a future that already happened

The honest test of any forecast: train it on the old years, then check it against years it never saw.

</div>

```python
# Stretch — train before 1990, test after 1990
train = df[df["year"] < 1990]
test  = df[df["year"] >= 1990]
for deg in [1, 3, 7]:
    c = np.polyfit(train["x"], train["anomaly"], deg)
    err = (np.polyval(c, test["x"]) - test["anomaly"]).abs().mean()
    print(f"degree {deg}: average miss on years it never saw = {err:.2f} degC")
# Which degree generalised best? Did the most flexible curve win?
```
