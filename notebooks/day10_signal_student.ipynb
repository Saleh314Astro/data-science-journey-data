```python
# ── The Daily Ritual — today, all four witnesses at once ─────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base = "https://raw.githubusercontent.com/Saleh314Astro/data-science-journey-data/main/clean/"
temp = pd.read_csv(base + "gistemp_monthly.csv",     parse_dates=["date"]).rename(columns={"value": "anomaly"}).dropna()   # NASA
co2  = pd.read_csv(base + "keeling_co2_monthly.csv",  parse_dates=["date"]).rename(columns={"value": "co2"}).dropna()       # NOAA
ice  = pd.read_csv(base + "nsidc_seaice_monthly.csv", parse_dates=["date"]).rename(columns={"value": "extent"}).dropna()    # NSIDC
sea  = pd.read_csv(base + "sealevel_monthly.csv",     parse_dates=["date"]).rename(columns={"value": "sealevel"}).dropna()  # CSIRO
september = ice[ice["date"].dt.month == 9]
print(len(temp), len(co2), len(september), len(sea), "rows -- four witnesses ready")
```

### Prediction 1 — four witnesses

- **Name:** …
- **My prediction:** …

```python
# The exhibit hall: four independent records, one figure
fig, axes = plt.subplots(2, 2, figsize=(13, 8))
axes[0,0].plot(temp["date"], temp["anomaly"].rolling(120, center=True).mean(), color="tab:red")
axes[0,0].axhline(0, color="gray", lw=0.7); axes[0,0].set_title("Temperature anomaly (degC) - NASA")
axes[0,1].plot(co2["date"], co2["co2"], color="seagreen"); axes[0,1].set_title("CO2 (ppm) - NOAA")
axes[1,0].plot(september["date"], september["extent"], color="tab:blue"); axes[1,0].set_title("September sea ice (million km2) - NSIDC")
axes[1,1].plot(sea["date"], sea["sealevel"], color="tab:purple"); axes[1,1].set_title("Sea level (mm) - CSIRO")
for ax in axes.flat: ax.set_xlabel("Year")
fig.suptitle("One Planet, Many Signals", fontsize=15)
fig.tight_layout()
plt.show()
```
